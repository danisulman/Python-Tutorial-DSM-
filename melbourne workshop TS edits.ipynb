# Numbers | Sequences

A common operation is to create a sequence. The range function achieves this and is commonly used in Python for iteration (eg for loops). Each element of the sequence is an integer, the collection is a list.

Let's look up the definition of range using `?range`.

Range produces a sequence of integers from start (inclusive) to stop (exclusive) by step. 

Step can be an increment or decrement (eg a countdown).

# Range object

**NOTE**: One difference between Python 2 and 3 is the `range` function. In Python 2, range() produces a list of the sequence (eager evaluation) while in Python 3 this is output as a range object (lazy evalation). As per Python documentation:

| The advantage of the range type over a regular list or tuple is that a range object will always take the same (small) amount of memory, no matter the size of the range it represents (as it only stores the start, stop and step values, calculating individual items and subranges as needed).

To access or print the sequence, use `list`, (eg `list(range(0,5))`) or use a slice (eg `range(0,5)[0]`)


# Range examples

Syntax: `range(start,stop,step)`

`list(range(0,5))` will evaluate the sequence and return `[0,1,2,3,4]`
`range(0,100000000)[100]` will (very quickly) return 99

Any element in a list can be accessed using with the required index specified in square brackets, eg:

# Pandas objects | Series

A series is a one dimensional array. (Think a column of data). The data values are one dimensional however we will see that the index can be multi-dimensional (a feature unique to Pandas).

The input to create a Series can be a list, Numpy array, or dictionary.

The first argument to series is the data. Index and Name are optional arguments.

```python
series(data=...)
series(data=data,name='name',index=index)
```

Index and name are optional. The data argument can be left unassigned.

If the input to Series is a dictionary, the index will the keys and the Series values will be the dictionary values.

## AirBnB rental price model (if time)
We'll build a model to predict the weekly price of a rental based on some features (number of bedrooms and type of property)

````python
# first we inspect the data - lets see the counts of properties in each local government area (note we've normalised the data)
airbnb['neighbourhood_cleansed'].value_counts(normalize = True, sort = True).head()
````

````python
# let's look at some predictors for the Melbourne local government area only

# select some predictors
predictors = ['bedrooms','property_type', 'price']
# filter data using query to see only Melbourne data
melb = airbnb.query("neighbourhood_cleansed == 'Melbourne'")[predictors]
# get the size of the data
print(melb.shape)
# inspect the predictors for Melbourne
melb.head()
````

````python
# for apartments in Melbourne, lets see what types of properties are more common (note data is normalised)
melb['property_type'].value_counts(normalize = True, sort = True).head()
````

````python
# above data but not normalised
melb['property_type'].value_counts(sort = True)
````

````python
# property_type has too many unique values. Lets throw away rows with rare values (at least 30) and convert the data to dummy version
# transform is an advanced Pandas aggregation method -- consider it as applying an aggregation and returning the aggregated result for each record.
melb_dummy = pd.get_dummies(melb)[melb.groupby('property_type')['property_type'].transform('size') > 30]

print('shape of the data frame is ',melb_dummy.shape)
# inspect the data - notice there are new columns corresponding to the property type with binary entries - this is what it means to use dummies
melb_dummy.head()
````

````python
# There may be missing/null prices
Xy = melb_dummy.dropna()         # remove missing values
y = Xy.price                     # define variable we're trying to predict
X = Xy.drop(["price"], axis=1)   # define features used to predict price
Xy.shape                         # inspect shape of data frame
X.head()                         # inspect some of the feature we'll use
````

# First model: decision tree
A decision tree uses a tree-like model of decisions and their possible consequences. It is one way to display an algorithm that only contains conditional control statements. It is a popular tool in machine learning.

For more information see:
```python
# python help
?DecisionTreeClassifier
```
or the Presciient course: Predictive analytics, machine learning, data science and AI

````python
from sklearn.tree import DecisionTreeClassifier 
tree = DecisionTreeClassifier(max_depth=3, min_samples_split=20, random_state=99)
tree.fit(X, y)
preds = pd.DataFrame({'y': y, 'tree': tree.predict(X)})
````

````python
# look at the predictions
# tree is the predicted price
# y is the actual price
preds.head(20)
````

# Second model: least squares regression

Regression uses a linear function to describe the data

For more information see the Presciient course: Predictive analytics, machine learning, data science and AI

We are ready to fit our first model - decision tree

````python
import statsmodels.api as sm
# build model
model = sm.OLS(y, X).fit()
# Print out the statistics
model.summary()
````

````python
# lets look at the predictions of this model (ols) with the decision tree predcitions (tree) and actual values (y)
preds['ols'] = model.predict(X)  # make the predictions by the model
preds.head()
````

## Plotting results
